# db/conexion.py

```python
import mysql.connector

# Configuración que coincide con tu XAMPP actual
DB_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "",  # Vacío por defecto en XAMPP [cite: 305]
    "database": "inventario",
    "port": 3306 [cite: 303]
}

def obtener_conexion():
    try:
        conexion = mysql.connector.connect(**DB_CONFIG)
        return conexion
    except mysql.connector.Error as err:
        print(f"Error de conexión: {err}")
        return None
```

# CategoriaDAO

```python
class CategoriaDAO:
    @staticmethod
    def insertar(categoria):
        conexion = obtener_conexion()
        if not conexion:
            return False

        try:
            cursor = conexion.cursor()
            # Consulta Parametrizada para seguridad [cite: 311, 313]
            sql = "INSERT INTO categorias (nombre, descripcion, activo) VALUES (%s, %s, %s)"
            valores = (categoria.nombre, categoria.descripcion, categoria.activo)

            cursor.execute(sql, valores)
            conexion.commit() # Guarda cambios en MariaDB [cite: 324, 384]
            categoria.id = cursor.lastrowid # Recupera el ID generado [cite: 330]
            return True
        except mysql.connector.Error as err:
            print(f"Error al insertar: {err}")
            conexion.rollback() # Revierte si hay error [cite: 331, 384]
            return False
        finally:
            cursor.close()
            conexion.close()

    @staticmethod # <--- Asegúrate de que este nivel coincida con el de arriba
    def actualizar(categoria):
        conexion = obtener_conexion()
        if not conexion:
            return False

        try:
            cursor = conexion.cursor()
            # El ID va de último para coincidir con el WHERE [cite: 341, 342]
            sql = "UPDATE categorias SET nombre=%s, descripcion=%s, activo=%s WHERE id=%s"
            valores = (categoria.nombre, categoria.descripcion, categoria.activo, categoria.id)

            cursor.execute(sql, valores)
            conexion.commit()
            return cursor.rowcount > 0
        except mysql.connector.Error as err:
            print(f"Error al actualizar: {err}")
            conexion.rollback()
            return False
        finally:
            cursor.close()
            conexion.close()
    @staticmethod
    def listar_todo():
        conexion = obtener_conexion()
        categorias = []
        if not conexion:
            return categorias

        try:
            cursor = conexion.cursor()
            cursor.execute("SELECT id, nombre, descripcion, activo FROM categorias")

            for (id, nombre, desc, activo) in cursor:
                cat = Categoria(nombre, desc, activo, id)
                categorias.append(cat)

            return categorias
        finally:
            cursor.close()
            conexion.close()
```

# ProductoDAO

```python
class ProductoDAO:
    @staticmethod
    def insertar(producto):
        conexion = obtener_conexion()
        if not conexion: return False
        try:
            cursor = conexion.cursor()
            # Consulta parametrizada (%s) para evitar Inyección SQL
            sql = "INSERT INTO productos (nombre, precio, stock, categoria_id) VALUES (%s, %s, %s, %s)"
            valores = (producto.nombre, producto.precio, producto.stock, producto.categoria_id)
            cursor.execute(sql, valores)
            conexion.commit() # Confirma la transacción
            return True
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            conexion.rollback() # Revierte si la FK falla o hay otro error
            return False
        finally:
            cursor.close()
            conexion.close()
    @staticmethod
    def buscar_por_nombre(termino):
        conexion = obtener_conexion()
        productos = []
        try:
            cursor = conexion.cursor()
            # El % es el comodín para buscar coincidencias parciales
            sql = "SELECT id, nombre, precio, stock, categoria_id FROM productos WHERE nombre LIKE %s"
            # Formateamos el término para que busque en cualquier parte del nombre
            cursor.execute(sql, (f"%{termino}%",))

            for (id, nom, pre, st, cat) in cursor:
                productos.append(Producto(nom, pre, st, cat, id))
            return productos
        finally:
            cursor.close()
            conexion.close()
```

# view/inventario_view.py

```python
class InventarioView:
    @staticmethod
    def mostrar_categorias(categorias):
        print("\n" + "="*50)
        print(f"{'ID':<5} | {'NOMBRE':<15} | {'ESTADO':<10}")
        print("-" * 50)

        if not categorias:
            print("No hay categorías registradas.")
        else:
            for cat in categorias:
                estado = "Activo" if cat.activo else "Inactivo"
                # :<5 alinea el ID a la izquierda con 5 espacios
                # :<15 alinea el nombre a la izquierda con 15 espacios
                print(f"{cat.id:<5} | {cat.nombre:<15} | {estado:<10}")
        print("="*50 + "\n")

    @staticmethod
    def capturar_datos_categoria():
        print("\n--- Nueva Categoría ---")
        nombre = input("Nombre: ")
        descripcion = input("Descripción: ")
        return nombre, descripcion
```

```python
class Categoria:
    def __init__(self, nombre, descripcion, activo=True, id=None):
        self.id = id  # El ID es None al inicio porque MariaDB lo asigna después [cite: 279, 280]
        self.nombre = nombre
        self.descripcion = descripcion
        self.activo = activo

    def __str__(self):
        # Retorna una cadena para mostrar la categoría fácilmente [cite: 277, 278]
        return f"{self.id} | {self.nombre} | {self.descripcion} | {'Activo' if self.activo else 'Inactivo'}"
```

# Producto

```python
class Producto:
    def __init__(self, nombre, precio, stock, categoria_id, id=None):
        self.id = id
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
        self.categoria_id = categoria_id  # Esta es la Clave Foránea (FK) [cite: 283, 285]

    def __str__(self):
        return f"{self.id} | {self.nombre} | ${self.precio} | Stock: {self.stock} | Cat: {self.categoria_id}"
```
